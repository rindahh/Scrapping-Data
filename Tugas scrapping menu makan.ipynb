```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://foodspot.co.id/cari/menu/snackbox/8,108,/1000/500000/0/2?restaurant_type=&searchby=menu&cat1=0&cat2=0&cat3=0&cat4=0&cat5=0&cat6=0&cat7=0&cat8=0&namakota=Jakarta+Selatan&namakecamatan=Tebet&kota=7&kecamatan=108&deliveryarea%5B0%5D=108&budgetmin=1000&budgetmax=99999999&sortby=2&page=1')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
foods = soup.find_all('div', class_='post-menu')
```

```python
foods
```

```python
name_foods = soup.find_all('div', class_='left_price search_result revision')
```

```python
name_foods
```

```python
data_menu = []
for name_food in name_foods:
    data_menu.append(name_food.text)
```

```python
brands = soup.find_all('div', class_='pull-left restaurant__search_result__002 verMobiles')
```

```python
brands
```

```python
data_brand = []
for brand in brands:
    data_brand.append(brand.text)
```

```python
prices = soup.find_all('div', class_='pull-right price__search_result__002')
```

```python
prices
```

```python
data_price = []
for price in prices:
    data_price.append(price.text)
```

```python
ratings = soup.find_all('div', class_='pull-left newRating side2020')
```

```python
ratings
```

```python
data_rating = []
for rating in ratings:
    data_rating.append(rating.text)
```

```python
data = {
    'name_foods' : data_menu,
    'brands' : data_brand,
    'prices' : data_price,
    'ratings' : data_rating
}
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel('Tugas Data Scrapping.xlsx')
```
